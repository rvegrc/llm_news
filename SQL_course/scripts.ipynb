```python
import psycopg2

from dotenv import load_dotenv
import os
# Load environment variables from .env file
load_dotenv()

import pandas as pd

# constants
PG_HOST = os.getenv('PG_HOST')
PG_PORT = os.getenv('PG_PORT')
PG_USER = os.getenv('PG_USER')
PG_PASS = os.getenv('PG_PASS')
PG_DB = os.getenv('PG_DB')

# Connect to the PostgreSQL database
def connect_to_db():
    try:
        connection = psycopg2.connect(
            host=PG_HOST,
            port=PG_PORT,
            user=PG_USER,
            password=PG_PASS,
            database=PG_DB
        )
        print("Connection to the database established successfully.")
        return connection
    except Exception as e:
        print(f"Error connecting to the database: {e}")
        return None

```

```python
def df_from_db(query:str, conn) -> pd.DataFrame:
    '''Create a DataFrame from a SQL query using a database connection.'''
    try:
        # Rollback in case a previous transaction failed
        conn.rollback()

        # Open a cursor to perform database operations
        cur = conn.cursor()

        # Execute query
        cur.execute(query)
        
        # Fetch all results from the executed query
        results = cur.fetchall()

        # Get column names from cursor description
        columns = [desc[0] for desc in cur.description]

        # Create DataFrame with column names
        df = pd.DataFrame(results, columns=columns)

        return df


    except Exception as e:
        print(f"Error: {e}")
        conn.rollback()

    finally:
        cur.close()
```

```python
conn = connect_to_db()

query = '''
            SELECT
                id,
                username,
                email,
                date_joined
            FROM users u
            ORDER BY date_joined desc, id asc
            '''

try:
    # Rollback in case a previous transaction failed
    conn.rollback()

    # Open a cursor to perform database operations
    cur = conn.cursor()

    # Execute query
    cur.execute(
        '''
        SELECT
            id,
            username,
            email,
            date_joined
        FROM users u
        ORDER BY date_joined desc, id asc
        '''
    )

    # Fetch all results from the executed query
    results = cur.fetchall()

    # Get column names from cursor description
    columns = [desc[0] for desc in cur.description]

    # Create DataFrame with column names
    df = pd.DataFrame(results, columns=columns)
    display(df)

except Exception as e:
    print(f"Error: {e}")
    conn.rollback()

finally:
    cur.close()
```
